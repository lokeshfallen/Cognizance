```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20,10))
x =  np.arange(0, 10, 0.1)
y = np.sin(x)
w = np.arange(0,10,0.1)
z = np.cos(x)
ax[0].plot(x, y, label = 'sinx')
ax[1].plot(w, z, label = 'cosx')
ax[2].plot(x, y, label = 'sinx')
ax[2].plot(w, z, label = 'cosx')

ax[0].legend()
ax[1].legend()
ax[2].legend()
```
