```python
import seaborn as sb
import pandas as pd
```

```python
df = pd.read_csv('ADANIPORTS.CSV')
```

```python
sb.histplot(df, x = df["Symbol"], hue= df["Symbol"], shrink = 0.6)
```

```python
bt  = sb.barplot(x = df["Symbol"], y = df["Turnover"], hue = df["Symbol"]) // To find the maximum turnover made by each port.
```

```python
sb.scatterplot(data = df, x = df["Open"], y = df["Close"])
```

```python
data = df.pivot(index = "Symbol", columns = "Date", values = "Turnover")
```

```python
ax = sb.heatmap(data)
```
