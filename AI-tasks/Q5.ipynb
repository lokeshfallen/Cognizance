```python
import csv
import matplotlib.pyplot as plt 
```

```python
with open('Q5.csv', 'w', newline = '') as f:
    csv_writer = csv.writer(f)
    for x in range(700, 2410, 10):
        y = 10*x^2 + 2*x
        csv_writer.writerow([x, y])
```

```python
arr1 = []
arr2 = []
with open('Q5.csv', 'r') as rf:
    data = csv.reader(rf, delimiter = ',')
    for row in data:
        arr1.append(int(row[0]) / 100)
        arr2.append(int(row[1])/1000)
```

```python
plt.plot(arr1, arr2)
plt.title('Task - 5')
plt.xlabel('size of houses (in 100)')
plt.ylabel('Price of the house (in 1000)')
```
